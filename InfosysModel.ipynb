```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

infosys = pd.read_csv(r'C:\Users\MY\Downloads\Datasets\Nifty50\INFY.csv')

infosys.head()
```

```python
infosys.info()
```

```python
from datetime import datetime

month = []
year = []

Date = infosys['Date']

for d in Date:
    d1 = datetime.strptime(d, '%Y-%m-%d')
    m = d1.month
    y = d1.year
    month.append(m)
    year.append(y)
```

```python
infosys['month'] = month
infosys['year'] = year
```

```python
infosys.head()
```

```python
infosys['monthly_avg'] = infosys.groupby(['year', 'month']).Close.transform('mean')
```

```python
infosys.head()
```

```python
infosys.drop_duplicates(subset = ['monthly_avg'], keep = 'first', inplace = True)
```

```python
infosys.head()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(infosys['monthly_avg'], model = 'additive')

fig = plt.figure()  
fig = result.plot()  
fig.set_size_inches(16, 9)
```

```python
infosysdiff = infosys['monthly_avg'] - infosys['monthly_avg'].shift(1)
infosysdiff2 = infosysdiff - infosysdiff.shift(12)
```

```python
plt.figure(figsize = (14, 6))

infosysdiff.plot()
plt.xlabel('Time')
plt.ylabel('Differenced Stock Prices d = 1')
```

```python
plt.figure(figsize = (14, 6))

infosysdiff2.plot()
plt.xlabel('Time')
plt.ylabel('Differenced Stock Prices D = 1')
```

```python
from statsmodels.tsa.stattools import kpss, adfuller

X = infosysdiff2.dropna().values
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
def kpss_test(timeseries):
    print("Results of KPSS Test:")
    kpsstest = kpss(timeseries, regression="c")
    kpss_output = pd.Series(
        kpsstest[0:3], index=["Test Statistic", "p-value", "Lags Used"]
    )
    for key, value in kpsstest[3].items():
        kpss_output["Critical Value (%s)" % key] = value
    print(kpss_output)
kpss_test(infosysdiff2.dropna().values)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(infosysdiff2.dropna().values, lags = 30)
```

```python
plot_pacf(infosysdiff2.dropna().values, lags = 30, method = 'ywm')
```

```python
import pmdarima as pm

model = pm.auto_arima(infosys['monthly_avg'], start_p = 0, start_q = 0,
                      max_p = 3, max_q = 3, # maximum p and q
                      m = 12,              # frequency of series
                      d = 1,
                      seasonal = True,
                      D = 1,
                      start_P = 0,
                      start_Q = 0,
                      max_P = 3,
                      max_Q = 3,
                      trace = True,
                      error_action = 'ignore',  
                      suppress_warnings = True, 
                      stepwise = True)

print(model.summary())
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model_sarimax = SARIMAX(infosys['monthly_avg'], order = (1, 1, 0), seasonal_order = (0, 1, 1, 12)) 
model_sarimax_fit = model_sarimax.fit()  
print(model_sarimax_fit.summary())
```

```python
n_periods = 24

fitted, confidence = model.predict(n_periods = n_periods, return_conf_int = True)
index_of_fc = pd.date_range(infosys['monthly_avg'].index[-1], periods = n_periods, freq = 'MS')


fitted_series = pd.Series(fitted, index = index_of_fc)
lower_series = pd.Series(confidence[ : , 0], index = index_of_fc)
upper_series = pd.Series(confidence[ : , 1], index = index_of_fc)

plt.figure(figsize = (14, 6))
plt.plot(infosys['monthly_avg'], label = 'Closing')
preds = model_sarimax_fit.predict(start = infosys['monthly_avg'].index[0], end = len(infosys['monthly_avg']))
preds.plot(label = 'predicted')
plt.plot(fitted_series, color = 'darkgreen', label = 'forecasted')
plt.fill_between(lower_series.index,
                lower_series,
                upper_series,
                color = 'k',
                alpha = 0.15)

plt.title("SARIMA - Final Forecast of Infosys Data")
plt.legend()
plt.show()
```
